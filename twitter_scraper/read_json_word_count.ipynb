# Read the file and assign the json data to a pandas dataframe

```python
idx = 0
words = []
for idx in df.index:
    [words.append(word) for word in df.text[idx].lower().split()]
```

```python
words_dict = {}
for word in set(words):
    words_dict.setdefault(word,  1)
```

```python
for word in words_dict.keys():
    words_dict[word] = words.count(word)
```

```python
import operator

sorted_x = sorted(words_dict.items(), key=operator.itemgetter(1), reverse = True )
```

```python
count_df = pd.DataFrame(sorted_x, columns = ['word','count'])
```
