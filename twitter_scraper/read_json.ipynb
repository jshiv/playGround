# import libraries

```python
import json
import pandas as pd
import numpy as np

filename = './results/tweets copy.json' #give the filename!
```

# Read the file and assign the json data to a pandas dataframe

```python
with open(filename, 'r') as f:
    l = f.readlines()

data = [json.loads(s) for s in l]
df = pd.DataFrame(data)
```
